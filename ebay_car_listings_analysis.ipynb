### Normalizing the brands popularity.

```python
autos["brand"].unique()
```

```python
autos["brand"].value_counts(normalize = True)
```

Since the values_count() function gives the distribution of the respective brands on the website, we can select the top 20 as a sample set to understand the effects of brand values in the current dataset.

The top 20 brands also represent data above 5% dominance of the current listings in terms of the count.

```python
brands_dist = autos["brand"].value_counts(normalize = True)
```

```python
popular_brands = brands_dist[brands_dist > .05].index
```

```python
popular_brands
```

As evident in the above index series, the most popular brands are volkswagen, bmw, opel,mercedes-benz, audi and ford.

_Hence these brands can be appropriately utilized to work upon in aggregation._


### Calculating the mean price.

```python
pop_brands_mean_price = {}

for b in popular_brands:
    autos_pop_brands = autos[autos["brand"] == b]
    autos_pop_brands_mean_price = autos_pop_brands["price"].mean()
    pop_brands_mean_price[b] = float(autos_pop_brands_mean_price)
pop_brands_mean_price
```

__The above data helps in realizing that popular brands by prices are in the order of:__

__audi > mercedes-benz > bmw > volkswagen > ford > opel__

__HIGHEST > HIGHER       > HIGH >    MEDIUM >  LOW > LOWEST__

### Calculating the mean mileage.

*Here the pop_brands_mean_price can be utilized to form a series object of the mileage data from the 'odometer_km' column.*

```python
bmp_series = pd.Series(pop_brands_mean_price)
print(bmp_series)
```

```python
df = pd.DataFrame(bmp_series, columns=['mean_price'])
df
```

```python
pop_brands_mean_mileage = {}
for b in popular_brands:
    pop_brands_mileage = autos[autos["brand"] == b]
    autos_pop_brands_mean_mileage = pop_brands_mileage["odometer_km"].mean()
    pop_brands_mean_mileage[b] = float(autos_pop_brands_mean_mileage)
pop_brands_mean_mileage
```

```python
bmm_series = pd.Series(pop_brands_mean_mileage)
```

```python
bmm_series
```

```python
df["mean mileage"] = bmm_series
```

```python
df.sort_values("mean_price", ascending = True)
```

_There is no correlation between the mean price and mean mileage for the respective dataset._
